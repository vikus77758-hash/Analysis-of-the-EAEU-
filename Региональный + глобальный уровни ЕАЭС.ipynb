```python
from typing import Dict, List

#словари


CATEGORIES: Dict[str, List[str]] = {

    "Кол-во_сюжет_экономический": [
        "рынок", "общий рынок", "единый рынок", "ёмкий рынок",
        "торговля", "товарооборот", "экспорт", "импорт", "поставки",
        "инвестиции", "инвесторы", "капитал", "капиталовложения",
        "ввп", "экономический рост", "рост экономики", "развитие экономики", "конкурентоспособность",
        "барьеры", "ограничения", "пошлины", "тарифы", "нетарифные барьеры",
        "производительность", "добавленная стоимость", "промышленность", "отрасли",
        "бизнес", "предприниматели", "компании", "предприятия",
        "интеграция рынков", "экономическая интеграция", "экономический союз",
    ],

    "Кол-во_сюжет_геополитический": [
 "центр силы", "полюс", "полюс роста", "центр мирового", "центр регионального",
        "многополярный мир", "многополярность",
        "суверенитет", "независимость", "стратегическая автономия",
        "геополитика", "геополитический", "геополитические интересы", "стратегические интересы",
        "мировой порядок", "международная система", "архитектура безопасности",
        "влияние", "лидерство", "позиционирование", "роль в мире",
        "противодействие", "баланс", "противостояние",
    ],

    "Кол-во_Сюжет_цивилизационный": [
         "цивилизация", "цивилизационный", "евразийская цивилизация",
        "общая история", "историческое единство", "исторические корни", "наследие",
        "культура", "культурные традиции", "духовные ценности", "духовность",
        "идентичность", "общая идентичность", "общность",
        "традиционные ценности", "общие ценности", "менталитет",
        "особый путь", "свой путь развития", "путь наших народов",
        "братские народы", "многовековые связи", "историческое братство",
    ],

    "Кол-во_сюжет_институциональный": [
"договор", "соглашение", "ратификация", "ратифицирован", "протокол", "решение", "документ", "орган", "комиссия", "совет", "комитет", "рабочая группа", "институт", "институты союза", "регламент", "процедура", "механизм", "инструменты", "нормативная база", "дорожная карта", "дорожные карты", "программа", "план мероприятий", "компетенции", "полномочия", "юрисдикция", "порядок председательства", "распределение голосов", "структура органов", "правовая база", "договорно-правовая база", "договорно‑правовая база", "юрисдикция суда",
    ],

    "Кол-во_сюжет_социально_гуманитарный": [
        "граждане", "наши граждане", "люди", "население",
        "трудовая миграция", "мигранты", "условия труда", "социальные гарантии",
        "образование", "университеты", "студенты", "академические обмены",
        "наука", "научное сотрудничество", "исследования",
        "культура", "культурные обмены", "язык", "гуманитарные проекты",
        "здравоохранение", "медицина", "фармацевтика", "социальная защита",
        "пенсии", "социальные выплаты", "уровень жизни", "качество жизни",
        "молодежь", "молодёжные программы", "молодежные программы", "студенческие программы",
    ],

    "Кол-во_cюжет_транспортно_логистический": [
     "автомобильный транспорт", "железнодорожный транспорт",
"авиасообщение", "морской транспорт", "мультимодальный транспорт",
"транспортная инфраструктура", "железная дорога", "пункт пропуска",
"логистика", "грузоперевозки", "транзит", "транзитный потенциал",
"грузооборот", "контейнерные перевозки",
"связанность", "транспортная связанность", "единое транспортное пространство",
"транспортный коридор", "шёлковый путь", "шелковый путь",
"север-юг", "транспортный маршрут",
"транспортные тарифы", "либерализация перевозок",
"цифровая логистика", "умная граница", "навигационные пломбы",
"логистический хаб", "сухой порт", "контейнерный терминал",

    ],
}

def normalize(text: str) -> str:
    """Нижний регистр + нормализация ё → е."""
    return text.lower().replace("ё", "е")


def count_occurrences(text: str, patterns: List[str]) -> int:
    """Подсчёт общего числа вхождений всех паттернов."""
    norm_text = normalize(text)
    total = 0
    for p in patterns:
        p_norm = normalize(p)
        if not p_norm:
            continue
        total += norm_text.count(p_norm)
    return total


def analyze_text(text: str):
    norm_text = normalize(text)

    # Подсчёт по категориям
    category_counts: Dict[str, int] = {}
    for category, patterns in CATEGORIES.items():
        category_counts[category] = count_occurrences(norm_text, patterns)

    # Ключевые корни
    evraz_count = norm_text.count("евраз")

    # Итог
    total = sum(category_counts.values())

    # Вывод
    print("=== Количество вхождений по категориям ===")
    for cat, cnt in category_counts.items():
        pct = (cnt / total * 100) if total > 0 else 0
        bar = "█" * int(pct / 2)
        print(f"  {cat}: {cnt}  ({pct:.1f}%)  {bar}")

    print("\n=== Ключевые корни ===")
    print(f"Кол-во упоминаний (евраз-): {evraz_count}")

    print("\n=== Вывод: что такое евразийность в данном тексте ===")
    if total == 0:
        print("Маркеры не обнаружены. Проверьте, что текст вставлен корректно.")
    else:
        ranked = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
        dominant = ranked[0]
        print(f"Доминирует: {dominant[0]} ({dominant[1]} марк., {dominant[1]/total*100:.0f}%)")
        others = [(c, n) for c, n in ranked[1:] if n > 0]
        if others:
            print("Дополнительные сюжеты:")
            for c, n in others:
                print(f"  • {c}: {n} ({n/total*100:.0f}%)")
        if category_counts.get("Кол-во_Сюжет_цивилизационный", 0) == 0:
            print("→ Цивилизационный нарратив отсутствует.")
        t = category_counts.get("Сюжет_транспортно_логистический", 0)
        if t > 3:
            print(f"→ Транспортно-логистический сюжет выражен ({t} марк.).")



if __name__ == "__main__":
    #вставка текста
    text = """


"""
    analyze_text(text)
```
