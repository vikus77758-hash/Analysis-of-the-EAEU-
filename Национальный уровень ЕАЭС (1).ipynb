```python
import re
from typing import Dict

EVRAZ_ROOT = "евраз"
EAES_TOKEN = "еаэс"


def normalize(text: str) -> str:
    """Нижний регистр + е -> е."""
    return text.lower().replace("ё", "е")


def count_root(text: str) -> int:
    """Считает вхождения корня 'евраз' как подстроки."""
    return normalize(text).count(EVRAZ_ROOT)


def count_eaes(text: str) -> int:
    """Считает вхождения 'ЕАЭС' как отдельного токена."""
    norm = normalize(text)
    pattern = r"(?<![а-яёa-z\d])" + re.escape(EAES_TOKEN) + r"(?![а-яёa-z\d])"
    return len(re.findall(pattern, norm))


def analyze_text(text: str) -> Dict[str, int]:
    """Возвращает словарь с подсчётом корня 'евраз-' и аббревиатуры 'ЕАЭС'."""
    evraz_count = count_root(text)
    eaes_count  = count_eaes(text)
    total       = evraz_count + eaes_count

    norm = normalize(text)
    found_words = re.findall(r"[а-яё]*евраз[а-яё]*", norm)

    print("=" * 45)
    print("  Анализ текста: евраз- и ЕАЭС")
    print("=" * 45)
    print(f"  Корень 'евраз-' (словоформы) : {evraz_count}")
    if found_words:
        unique = sorted(set(found_words))
        print(f"  Найденные слова             : {', '.join(unique)}")
    print(f"  Аббревиатура 'ЕАЭС'         : {eaes_count}")
    print("-" * 45)
    print(f"  Итого упоминаний            : {total}")
    print("=" * 45)

    return {
        "евраз_корень": evraz_count,
        "ЕАЭС":         eaes_count,
        "итого":        total,
    }

#вставка текста
if __name__ == "__main__":
    text = """

    """

    results = analyze_text(text)
```
